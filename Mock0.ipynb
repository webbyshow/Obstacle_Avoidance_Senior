# DenseOpticalFlow

```python
import numpy as np
import cv2
import time



def draw_flow(img, flow, step=16):

    h, w = img.shape[:2]
    y, x = np.mgrid[step/2:h:step, step/2:w:step].reshape(2,-1).astype(int)
    fx, fy = flow[y,x].T

    lines = np.vstack([x, y, x-fx, y-fy]).T.reshape(-1, 2, 2)
    lines = np.int32(lines + 0.5)

    img_bgr = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    cv2.polylines(img_bgr, lines, 0, (0, 255, 0))

    for (x1, y1), (_x2, _y2) in lines:
        cv2.circle(img_bgr, (x1, y1), 1, (0, 255, 0), -1)

    return img_bgr


def draw_hsv(flow):

    h, w = flow.shape[:2]
    fx, fy = flow[:,:,0], flow[:,:,1]

    ang = np.arctan2(fy, fx) + np.pi
    v = np.sqrt(fx*fx+fy*fy)

    hsv = np.zeros((h, w, 3), np.uint8)
    hsv[...,0] = ang*(180/np.pi/2)
    hsv[...,1] = 255
    hsv[...,2] = np.minimum(v*4, 255)
    bgr = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

    return bgr




cap = cv2.VideoCapture(0)

suc, prev = cap.read()
prevgray = cv2.cvtColor(prev, cv2.COLOR_BGR2GRAY)


while True:

    suc, img = cap.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # start time to calculate FPS
    start = time.time()


    flow = cv2.calcOpticalFlowFarneback(prevgray, gray, None, 0.5, 3, 15, 3, 5, 1.2, 0)
    
    prevgray = gray


    # End time
    end = time.time()
    # calculate the FPS for current frame detection
    fps = 1 / (end-start)

    print(f"{fps:.2f} FPS")

    cv2.imshow('flow', draw_flow(gray, flow))
    cv2.imshow('flow HSV', draw_hsv(flow))


    key = cv2.waitKey(20) & 0xFF
    if key == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```

## Keypoint and Descriptor From ORB

```python
import cv2
import os

# Set the paths for the Left and Right folders
left_folder = os.path.abspath("Left5")
right_folder = os.path.abspath("Right5")

# Get sorted lists of image files from both folders
left_images = sorted([f for f in os.listdir(left_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
right_images = sorted([f for f in os.listdir(right_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])

# Ensure both folders have the same number of images
if len(left_images) != len(right_images):
    print("The number of images in the Left and Right folders does not match.")
else:
    for i, (left_img, right_img) in enumerate(zip(left_images, right_images)):
        # Read images from both folders
        left_path = os.path.join(left_folder, left_img)
        right_path = os.path.join(right_folder, right_img)

        left_frame = cv2.imread(left_path)#, cv2.IMREAD_GRAYSCALE)  # Convert to grayscale
        right_frame = cv2.imread(right_path)#, cv2.IMREAD_GRAYSCALE)

        # Check if images are loaded properly
        if left_frame is None or right_frame is None:
            print(f"Error loading images: {left_img}, {right_img}")
            continue

        # Initialize ORB detector
        orb = cv2.ORB_create()

        # Detect keypoints and compute descriptors
        keypoints_left, descriptors_left = orb.detectAndCompute(left_frame, None)
        keypoints_right, descriptors_right = orb.detectAndCompute(right_frame, None)

        # Use BFMatcher to match descriptors
        bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
        matches = bf.match(descriptors_left, descriptors_right)

        # Sort matches by distance
        matches = sorted(matches, key=lambda x: x.distance)

        # Draw matches
        matched_frame = cv2.drawMatches(left_frame, keypoints_left,
                                        right_frame, keypoints_right,
                                        matches, None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

        # Display the matched frame without saving
        cv2.imshow("Feature Matching (Left | Right)", matched_frame)

        # Wait for a key press
        key = cv2.waitKey(30)  # Adjust delay for playback speed
        if key == ord('q'):  # Press 'q' to exit
            break

    # Close all OpenCV windows
    cv2.destroyAllWindows()
    cv2.waitKey(1)

    print("Feature matching completed.")
```

### YOLO11 + Feature Matching + No Saving

```python
import cv2
import os
from ultralytics import YOLO

# Load the YOLOv11 model (replace 'yolo11s.pt' with your desired model size)
model = YOLO('yolo11s.pt')

# Set the paths for the Left and Right folders
left_folder = "Left5"
right_folder = "Right5"

# Get sorted lists of image files from both folders
left_images = sorted(os.listdir(left_folder))
right_images = sorted(os.listdir(right_folder))

# Ensure both folders have the same number of images
if len(left_images) != len(right_images):
    print("The number of images in the Left and Right folders does not match.")
else:
    # Initialize ORB detector
    orb = cv2.ORB_create()

    # Loop through image pairs
    for left_img, right_img in zip(left_images, right_images):
        # Read images from both folders
        left_path = os.path.join(left_folder, left_img)
        right_path = os.path.join(right_folder, right_img)

        left_frame = cv2.imread(left_path)
        right_frame = cv2.imread(right_path)

        # Apply YOLOv11 detection on both frames
        left_results = model(left_frame)
        right_results = model(right_frame)

        # Extract bounding boxes for both frames
        left_boxes = left_results[0].boxes.xyxy.cpu().numpy()
        right_boxes = right_results[0].boxes.xyxy.cpu().numpy()

        # Detect keypoints and descriptors for the whole frame
        keypoints_left, descriptors_left = orb.detectAndCompute(left_frame, None)
        keypoints_right, descriptors_right = orb.detectAndCompute(right_frame, None)

        # Function to filter keypoints **inside bounding boxes only**
        def filter_keypoints_in_boxes(keypoints, boxes):
            filtered_keypoints = []
            for kp in keypoints:
                x, y = kp.pt
                for box in boxes:
                    x1, y1, x2, y2 = box
                    if x1 <= x <= x2 and y1 <= y <= y2:
                        filtered_keypoints.append(kp)
                        break  # Stop checking once a keypoint is inside a box
            return filtered_keypoints

        # Filter keypoints based on bounding boxes
        filtered_keypoints_left = filter_keypoints_in_boxes(keypoints_left, left_boxes)
        filtered_keypoints_right = filter_keypoints_in_boxes(keypoints_right, right_boxes)

        # Compute descriptors again for the filtered keypoints
        def compute_descriptors(frame, keypoints):
            if len(keypoints) > 0:
                keypoints, descriptors = orb.compute(frame, keypoints)
                return keypoints, descriptors
            return [], None

        filtered_keypoints_left, descriptors_left = compute_descriptors(left_frame, filtered_keypoints_left)
        filtered_keypoints_right, descriptors_right = compute_descriptors(right_frame, filtered_keypoints_right)

        # Draw the bounding boxes on both frames
        for box in left_boxes:
            x1, y1, x2, y2 = map(int, box)
            cv2.rectangle(left_frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

        for box in right_boxes:
            x1, y1, x2, y2 = map(int, box)
            cv2.rectangle(right_frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Perform feature matching only if descriptors exist
        if descriptors_left is not None and descriptors_right is not None and len(descriptors_left) > 0 and len(descriptors_right) > 0:
            # Use BFMatcher for feature matching
            bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
            matches = bf.match(descriptors_left, descriptors_right)
            matches = sorted(matches, key=lambda x: x.distance)

            # Draw feature matches only for keypoints inside bounding boxes
            matched_frame = cv2.drawMatches(left_frame, filtered_keypoints_left,
                                            right_frame, filtered_keypoints_right,
                                            matches[:50], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

            # Display the frames side by side
            cv2.imshow("YOLO Detection + Feature Matching (Inside Bounding Box Only)", matched_frame)

        else:
            print("No sufficient keypoints inside bounding boxes for matching.")
            combined_frame = cv2.hconcat([left_frame, right_frame])
            cv2.imshow("Bounding Boxes (No Matches)", combined_frame)

        # Exit on 'q' key press
        key = cv2.waitKey(30)
        if key == ord('q'):
            break

    # Close all windows after completion
    cv2.destroyAllWindows()
    cv2.waitKey(1)
```

```python
import cv2
import os

# Set the paths for the Left and Right folders
left_folder = os.path.abspath("Left5")
right_folder = os.path.abspath("Right5")

# Select the desired descriptor: "SIFT", "BRISK", "AKAZE"
#descriptor_choice = "SIFT"  # Change to "BRISK" or "AKAZE" to test different descriptors
#descriptor_choice = "BRISK"  # Change to "BRISK" or "AKAZE" to test different descriptors
descriptor_choice = "AKAZE"  # Change to "BRISK" or "AKAZE" to test different descriptors

# Initialize the selected feature extractor and descriptor
if descriptor_choice == "SIFT":
    feature_extractor = cv2.SIFT_create()
    matcher = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)  # L2 Norm for float descriptors
elif descriptor_choice == "BRISK":
    feature_extractor = cv2.BRISK_create()
    matcher = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)  # Hamming distance for binary descriptors
elif descriptor_choice == "AKAZE":
    feature_extractor = cv2.AKAZE_create()
    matcher = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)  # Hamming for binary descriptors
else:
    raise ValueError("Invalid descriptor choice! Use 'SIFT', 'BRISK', or 'AKAZE'.")

# Get sorted lists of image files from both folders
left_images = sorted([f for f in os.listdir(left_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
right_images = sorted([f for f in os.listdir(right_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])

# Ensure both folders have the same number of images
if len(left_images) != len(right_images):
    print("The number of images in the Left and Right folders does not match.")
else:
    for i, (left_img, right_img) in enumerate(zip(left_images, right_images)):
        # Read images from both folders
        left_path = os.path.join(left_folder, left_img)
        right_path = os.path.join(right_folder, right_img)

        left_frame = cv2.imread(left_path, cv2.IMREAD_GRAYSCALE)  # Convert to grayscale
        right_frame = cv2.imread(right_path, cv2.IMREAD_GRAYSCALE)

        # Check if images are loaded properly
        if left_frame is None or right_frame is None:
            print(f"Error loading images: {left_img}, {right_img}")
            continue

        # Detect keypoints and compute descriptors using the selected method
        keypoints_left, descriptors_left = feature_extractor.detectAndCompute(left_frame, None)
        keypoints_right, descriptors_right = feature_extractor.detectAndCompute(right_frame, None)

        # Ensure descriptors exist
        if descriptors_left is None or descriptors_right is None:
            print("No descriptors found.")
            continue

        # Use the chosen matcher to find feature matches
        matches = matcher.match(descriptors_left, descriptors_right)
        matches = sorted(matches, key=lambda x: x.distance)

        # Draw the matches
        matched_frame = cv2.drawMatches(left_frame, keypoints_left,
                                        right_frame, keypoints_right,
                                        matches[:50], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

        # Display the matched frame
        cv2.imshow(f"Feature Matching ({descriptor_choice})", matched_frame)

        # Wait for a key press
        key = cv2.waitKey(30)
        if key == ord('q'):
            break

    # Close all OpenCV windows after completion
    cv2.destroyAllWindows()
    print(f"Feature matching completed using {descriptor_choice}.")
```
